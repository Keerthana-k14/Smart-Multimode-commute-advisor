# Phase 1: Exploratory Data Analysis (EDA)

In this phase, we analyze the Bangalore Traffic Dataset to understand the distributions and patterns of travel times and congestion levels.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

DATA_PATH = "../backend/data/raw/Banglore_traffic_Dataset.csv"
df = pd.read_csv(DATA_PATH)

print(f"Total records: {len(df)}")
df.head()
```

## 1. Distribution of Congestion Levels

Understanding the general traffic intensity in the city.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Congestion Level'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Congestion Levels in Bangalore')
plt.xlabel('Congestion Level (%)')
plt.ylabel('Frequency')
plt.show()
```

## 2. Average Travel Time Index by Area

Which areas are the most congested?

```python
plt.figure(figsize=(12, 6))
df.groupby('Area Name')['Travel Time Index'].mean().sort_values().plot(kind='barh', color='salmon')
plt.title('Average Travel Time Index by Area')
plt.xlabel('Mean Travel Time Index')
plt.show()
```
