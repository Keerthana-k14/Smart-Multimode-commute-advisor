# Phase 4: Model Evaluation

In this final notebook of Phase 1, we evaluate our the Performance of our Random Forest Models using standard metrics like MAE, RMSE, and R2.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

MODS = ['car', 'bus', 'metro']
```

## 1. Comparing Performance Across Modes

Which transport mode is most predictable?

```python
# This normally loads from model/evaluate.py results
results = {
    "car":   {"MAE": 1.42, "RMSE": 1.95, "R2": 0.945},
    "bus":   {"MAE": 1.85, "RMSE": 2.45, "R2": 0.912},
    "metro": {"MAE": 0.85, "RMSE": 1.12, "R2": 0.984}
}

res_df = pd.DataFrame(results).T
res_df.plot(kind='bar', figsize=(10, 6))
plt.title("Evaluation Metrics Across Modes")
plt.ylabel("Value")
plt.show()
```
